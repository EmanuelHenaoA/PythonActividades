```python
import pandas as pd
import numpy as np
import matplotlib as mpl
# 1. Leer la URL desde github del archivo llamado clientes.csv y conocer su informacion principal mediante el uso de atributos y metodos necesarios de pandas.
urlClientes = 'https://raw.githubusercontent.com/rubisabel00/ciencias-de-datos-Sena/main/DocumentosNB03/clientes.csv'
data_clientes = pd.read_csv(urlClientes, encoding='UTF=8', index_col=0)
```

```python
# 2. Realizar una descripcion estadistica de la informacion encontrada y normalizarla (Eliminar los null).

normali = data_clientes.dropna()
normali
```

```python

# 3. Generar una grafica representativa de la informacion obtenida.

print(normali.plot())
```

```python
# 4. Generar dos consultas de agrupamiento de valor agregado a la informacion obtenida.

normali.groupby('Frecuencia de compra').count()

normali.groupby('Frecuencia de compra').agg({'Nombre': "count"})
```

```python
# 5. Generar nuevas columnas segun sean requeridas para completar la informacion obtenida.

normali['Información'] = normali['Nombre'] + " Su categoria de productos es " + normali['Categoría de productos comprados']
normali
```
